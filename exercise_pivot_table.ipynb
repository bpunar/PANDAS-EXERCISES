PIVOT

```python
import pandas as pd
```

```python
d = {'date': ['5/1/2017', '5/2/2017', '5/3/2017', '5/1/2017','5/2/2017','5/3/2017', '5/1/2017', '5/2/2017','5/3/2017'], 
     'city': ["new york", "new york", "new york", "mumbai", "mumbai", "mumbai", "beijing", "beijing", "beijing"], 
     'temparature' : [65, 66, 68, 75, 78, 82, 80, 77, 79],
     'humidity': [56, 58, 60, 80, 83, 85, 26, 30, 35]}
df = pd.DataFrame(data=d)
df
```

```python
df.pivot(index="date", columns="city")
```

```python
df.pivot(index="date", columns="city", values="humidity")
```

```python
df.pivot(index="date", columns="city", values="temparature")
```

```python
df.pivot(index="humidity", columns="city")
```

```python
d2 = {'date': ['5/1/2017', '5/1/2017', '5/2/2017', '5/2/2017','5/1/2017','5/1/2017', '5/2/2017', '5/2/2017'], 
     'city': ["new york", "new york", "new york", "new york", "mumbai", "mumbai", "mumbai", "mumbai"], 
     'temparature' : [65, 61, 70, 72,75, 78, 82, 80],
     'humidity': [56, 54, 60, 62, 80, 83, 85, 26]}
df = pd.DataFrame(data=d2)
df
```

```python
df.pivot_table(index="city",columns="date")
```

```python
df.pivot_table(index="city",columns="date", aggfunc="sum")
```

```python
df.pivot_table(index="city",columns="date", aggfunc="count")
```

```python
df.pivot_table(index="city",columns="date", aggfunc="diff")
```

```python
df.pivot_table(index="city",columns="date", aggfunc="mean")
```

```python
df.pivot_table(index="city",columns="date", margins=True)
```

```python
df.pivot_table(index="city",columns="date", margins=False)
```

```python
d3 = {'date': ['5/1/2017', '5/2/2017', '5/3/2017', '12/1/2017','12/2/2017','12/3/2017'], 
     'city': ["new york", "new york", "new york", "new york", "new york", "new york"], 
     'temparature' : [65, 61, 70, 30, 28, 25],
     'humidity': [56, 54, 60, 50, 52, 51]}
df = pd.DataFrame(data=d3)
df
```

```python
print(type(df['date']))
print(type(df['date'][0]))
```

```python
df['date'] = pd.to_datetime(df['date'])
print(type(df['date']))
print(type(df['date'][0]))
```

```python
df.pivot_table(index=pd.Grouper(freq="M", key="date"), columns='city')
```

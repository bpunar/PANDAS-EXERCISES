### Conditional Selection

An important feature of pandas is conditional selection using bracket notation, very similar to numpy:

```python
df3
```

```python
np.random.seed(101)
df3 = pd.DataFrame(randn(5,4), index = 'A B C D E'.split(), columns = 'W X Y Z'.split())
df3
```

```python
df3
```

```python
df3>2
```

```python
df3[df3>2]
```

```python
df3[df3["Z"]>0.5]
```

```python
df3[df3["Z"]>0.5]["Y"] #Series Form
```

```python
df3[df3["Z"]>0.5][["Y"]] #Dataframe form
```

```python
df3[df3["Z"]>0.5][["Y","Z"]] # Here it shows the Y and Z columns.
```

```python
df3[df3["Z"]>0.5]["Y"],["Z"] # It doesn't show the Z column here.
```

#### For two conditions you can use **|**  `or`,  **&**   `and` with parenthesis:

```python
df3[(df3['W']>0) & (df3['Y']<1)]
```

```python
df3[(df3['W']>0) & (df3['Y']<1)] = 100
df3
```

### Conditional selection using ``.loc[]`` and ``.iloc[]``

```python
np.random.seed(101)
df3 = pd.DataFrame(randn(5,4), index = 'A B C D E'.split(), columns = 'W X Y Z'.split())
df3
```

```python
df3.loc[(df3.W>0), ["X","Y"]]
```

```python
df3.loc[((df3.W>1) | (df3.Y<1)), ['Y','Z']]
```

## Multi-Index and Index Hierarchy

Let us go over how to work with Multi-Index, first we'll create a quick example of what a Multi-Indexed DataFrame would look like:

```python
# Index Levels
outside = ['M1', 'M1', 'M1', 'M2', 'M2', 'M2','M3', 'M3', 'M3']
inside = [1, 2, 3, 1, 2, 3, 5, 6, 7]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index=pd.MultiIndex.from_tuples(multi_index)
```

```python
df=pd.DataFrame(np.random.randn(9,4), index = hier_index, columns=['A','B','C','D'])
df
```

```python
df.loc["M1"]
```

```python
df.loc["M1"].loc[2]
```

```python
df.loc["M1"].loc[[2]]
```

```python
df.index
```

```python
df.index.names
```

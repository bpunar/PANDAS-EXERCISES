```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
from functools import reduce
import warnings
warnings.filterwarnings('ignore') 
```

```python
df1=pd.read_csv("continents2.csv")
df1
```

```python
df1=df1[["name","region", "sub-region"]]
df1
```

```python
df2=pd.read_csv("Tokyo Medals 2021.csv")
df2
```

```python
df1 = df1.rename(columns={"name":"Country"})
```

```python
df_merge1 = pd.merge(df2, df1, how='inner')
df_merge1
```

```python
df3=pd.read_csv("Poulation.csv")
df3 = df3.loc[0:216,:][["Country Name","2020 [YR2020]"]]
df3 = df3.rename(columns={"Country Name":"Country"})
df3 = df3.rename(columns={"Country Name":"Country", "2020 [YR2020]":"POPULATION 2020"})
df3
```

```python
df_merge2 = pd.merge(df_merge1, df3, how='inner')
df_merge2
```

```python
df4=pd.read_csv("life_expectancy.csv")
df4 = df4.loc[0:216,:][["Country Name","2019 [YR2019]"]]
df4 = df4.rename(columns={"Country Name":"Country", "2019 [YR2019]":"LIFE EXPECTANCY 2019"})
df4
```

```python
df_merge3 = pd.merge(df_merge2, df4, how='inner')
df_merge3
```

```python
df5=pd.read_csv("GDP_per_capita.csv")
df5 = df5.loc[0:216,:][["Country or Area","Value"]]
df5 = df5.rename(columns={"Country or Area":"Country", "Value":"GDP per CAPITA 2020"})
df5
```

```python
df_merge4 = pd.merge(df_merge3, df5, how='inner')
df_merge4.sample(10)
```

```python
df_merge4.info()
```

```python
df_merge4["LIFE EXPECTANCY 2019"] = df_merge4["LIFE EXPECTANCY 2019"].replace("..",0)
df_merge4["LIFE EXPECTANCY 2019"] = pd.to_numeric(df_merge4["LIFE EXPECTANCY 2019"], errors='coerce')
df_merge4["LIFE EXPECTANCY 2019"]  = df_merge4["LIFE EXPECTANCY 2019"] .round(1)
df_merge4['GDP per CAPITA 2020'] = df_merge4['GDP per CAPITA 2020'].round(0).astype(int)
df_merge4
```

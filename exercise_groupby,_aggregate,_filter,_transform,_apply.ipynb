```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.RandomState(0) # I don't understand why this part is not working.
df = pd.DataFrame({'key' : ['A','B', 'C', 'A', 'B' 'C'],
                   'data1' : range(6),
                   'data2' :rng.randint(0, 10, 6)},
                  columns = ['key','data1','data2'])
df
```

```python
d = {'key': ['A', 'B', 'C', 'A','B','C'], 'data1': [0, 1, 2, 3, 4, 5], 'data2' : [5, 0, 3, 3 ,7, 9]}
df = pd.DataFrame(data=d)
df
```

```python
df.groupby('key') # 
```

```python
df.groupby('key').aggregate('min') # If I don't specifically specify a column, it applies the function to all columns.
```

```python
df.groupby('key').aggregate('min','max') # Only the first function works here. others will not work. because ( ) is used. Indeed [ ] should have been used.
```

```python
df.groupby('key').aggregate(['min','max']) # Both functions work.
```

```python
#df.groupby('key').aggregate('min','max', np.mean) # [] should not forget
df.groupby('key').aggregate(['min','max', np.mean])
```

```python
df.groupby('key').aggregate({'data1' : 'min', # A different aggregate function can be applied to each column.
                             'data2' : 'max'})
```

```python
df.groupby('key').aggregate({'data1' : 'count', # A different aggregate function can be applied to each column.
                             'data2' : 'max'})
```

```python
df.groupby('key').aggregate({'data1' : 'count', # A different aggregate function can be applied to each column.
                             'data2' : np.mean})
```

```python
df.groupby('key').aggregate({'data1' : np.std, # A different aggregate function can be applied to each column.
                             'data2' : np.mean})
```

```python
# function is TRUE only B and C, so A is FALSE
def filter_func(x):
  return x['data2'].std() > 4
```

```python
df.groupby('key').transform(lambda x: x-x.mean())
# mean
# data1: A:1,5 B: 2.5, C:3.5
# data2: A:, B:3.5 C: 6 

#apply x-x.mean
# 0: data1 : (0-1.5=1.5), data2 = (5-4=1)
# 1: data1 : (1-2.5 =-1.5), data2=(0-3.5=-3.5)
#
```

```python
def norm_by_data2(x):
  x['data1'] /= x['data2'].sum()
# x['data1'] = X['data1'] / x['data2'].sum()
  return x
```

```python
df.groupby('key').apply(norm_by_data2)
```

```python
display(df, df.groupby('key').apply(norm_by_data2))
```
